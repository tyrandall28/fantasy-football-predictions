First load the desired seasons of player data.

```python
import pandas as pd
import os

# --- Configuration ---
DATA_DIR = '../data_collection/player_data_csvs'

# Using 2013-2023 for training to predict for 2024.
SEASONS_TO_LOAD = range(2013, 2024)


# --- Data Loading ---
all_season_dfs = []
print(f"Looking for data in: {os.path.abspath(DATA_DIR)}")

for season in SEASONS_TO_LOAD:
    file_path = os.path.join(DATA_DIR, f'fantasy_{season}.csv')
    
    if os.path.exists(file_path):
        print(f"Loading data for {season} season...")
        try:
            season_df = pd.read_csv(file_path)
            season_df['Season'] = season # Add a 'Season' column to identify the year for each row ( we are adding all seasons to the same dataframe)
            all_season_dfs.append(season_df)
        except Exception as e:
            print(f"Error loading {file_path}: {e}")
    else:
        print(f"Warning: File not found for season {season} at {file_path}")

# --- Combine and Inspect ---
if all_season_dfs:
    master_df = pd.concat(all_season_dfs, ignore_index=True)
    
    print("\\nSuccessfully loaded and combined data.")
    print(f"Seasons loaded: {sorted(master_df['Season'].unique())}")
    print(f"Total rows: {len(master_df)}")
    
    print('\\n--- First 5 Rows ---')
    display(master_df.head()) # quick view for sanity check, check season is first season in the defined range (i.e. 2013)

    print('\\n--- Last 5 Rows ---')
    display(master_df.tail()) # second quick view for consistency, check season is last season in the defined range (i.e. 2023)
    
    print('\\n--- DataFrame Info ---')
    master_df.info() # check column names and data types
else:
    print("\\nError: No data was loaded. Check file paths and season range.") 
```

Now clean the data. Need to remove NaNs and non-numerics like symbols. 
NaNs will be converted to 0s as this typically means a player did not record any activity in this category.

```python
# This code assumes 'master_df' exists from the previous cell.
import numpy as np

# --- Data Cleaning and Type Conversion ---

# 1. Clean Player Names
# Remove special characters like '*' (Pro-Bowl) and '+' (All-Pro) for consistent naming.
if 'player' in master_df.columns:
    master_df['player'] = master_df['player'].str.replace(r'[\\*\\+]', '', regex=True).str.strip()
    print("Cleaned player names.")

# 2. Identify Column Types
ID_COLS = ['player', 'team', 'position', 'player_id'] # text or identifier cols, all others are numeric
NUMERIC_COLS = [col for col in master_df.columns if col not in ID_COLS]

# 3. Convert Columns to Numeric and Handle Missing Values
# Coerce errors, turning any non-numeric value into NaN, then fill with 0.
master_df[NUMERIC_COLS] = master_df[NUMERIC_COLS].apply(pd.to_numeric, errors='coerce').fillna(0)
print(f"Converted {len(NUMERIC_COLS)} columns to numeric and filled NaNs with 0.")

# 4. Convert float columns that should be integers to int type
# These are columns representing counts, ranks, or whole numbers.
INT_COLS = [
    'rank', 'age', 'games', 'games_started', 'completions', 'pass_attempts', 
    'pass_yards', 'pass_tds', 'ints', 'rush_attempts', 'rush_yards', 'rush_tds',
    'targets', 'receptions', 'rec_yds', 'rec_tds', 'fumbles', 'fumbles_lost',
    'rush_and_rec_td_total', 'two_point_convs_made', 'two_point_conv_passes',
    'vbd', 'pos_rank', 'overall_rank', 'Season'
]
# Convert columns that actually exist in the DataFrame
existing_int_cols = [col for col in INT_COLS if col in master_df.columns]
master_df[existing_int_cols] = master_df[existing_int_cols].astype(np.int64)
print(f"Converted {len(existing_int_cols)} columns to integer type.")

# --- Verification ---
print("\\n--- DataFrame Info After Cleaning ---")
master_df.info()

print("\\n--- Sample of Cleaned Data ---")
# display(master_df.head())
print(master_df.head())
```

Feature engineering next

```python
# This code assumes 'master_df' exists and is cleaned from the previous cells.
import pandas as pd
import numpy as np

# --- Configuration ---
N_LAG_SEASONS = 3 # Set to 3 for now to keep feature count manageable, but can be increased

STATS_TO_LAG = [
    'age', 'games', 'games_started', 'completions', 'pass_attempts', 
    'pass_yards', 'pass_tds', 'ints', 'rush_attempts', 'rush_yards', 
    'yards_per_rush_attempt', 'rush_tds', 'targets', 'receptions', 'rec_yds', 
    'yards_per_rec', 'rec_tds', 'fumbles', 'fumbles_lost',
    'ppr_fantasy_points'
]
TARGET_VARIABLE = 'ppr_fantasy_points'

# --- Feature Engineering (Enhanced) ---
print("Starting ENHANCED feature engineering...")
master_df = master_df.sort_values(by=['player_id', 'Season'])
all_features = []
grouped = master_df.groupby('player_id')

for player_id, player_df in grouped:
    for i in range(1, len(player_df)):
        target_season_data = player_df.iloc[i]
        
        features = {
            'player_id': player_id,
            'player_name': target_season_data['player'],
            'position': target_season_data['position'],
            'prediction_season': int(target_season_data['Season']),
            'target_ppr_points': target_season_data[TARGET_VARIABLE]
        }

        # Step 1: Create basic lagged features (S-1, S-2, etc.)
        for lag in range(1, N_LAG_SEASONS + 1):
            if i - lag >= 0:
                past_season = player_df.iloc[i - lag]
                for stat in STATS_TO_LAG:
                    features[f'{stat}_s_minus_{lag}'] = past_season[stat]
            else:
                for stat in STATS_TO_LAG:
                    features[f'{stat}_s_minus_{lag}'] = 0
        
        # Step 2: Create Derived Features (Averages, Trends, etc)
        # --- Per-Game & Efficiency Metrics ---
        for lag in range(1, N_LAG_SEASONS + 1):
            games_played = features.get(f'games_s_minus_{lag}', 0)
            if games_played > 0:
                features[f'ppr_per_game_s_minus_{lag}'] = features[f'ppr_fantasy_points_s_minus_{lag}'] / games_played
                features[f'rush_yards_per_game_s_minus_{lag}'] = features[f'rush_yards_s_minus_{lag}'] / games_played
                features[f'rec_yards_per_game_s_minus_{lag}'] = features[f'rec_yds_s_minus_{lag}'] / games_played
                features[f'pass_yards_per_game_s_minus_{lag}'] = features[f'pass_yards_s_minus_{lag}'] / games_played
            else:
                features[f'ppr_per_game_s_minus_{lag}'] = 0
                features[f'rush_yards_per_game_s_minus_{lag}'] = 0
                features[f'rec_yards_per_game_s_minus_{lag}'] = 0
                features[f'pass_yards_per_game_s_minus_{lag}'] = 0

        # --- Multi-Season Aggregates & Volatility ---
        # Last 2 Seasons
        last_2_seasons_points = [features.get(f'ppr_fantasy_points_s_minus_{j}', 0) for j in range(1, 3)]
        valid_points_l2 = [p for p in last_2_seasons_points if p > 0]
        if valid_points_l2:
            features['avg_ppr_last_2_seasons'] = np.mean(valid_points_l2)
        else:
            features['avg_ppr_last_2_seasons'] = 0
            
        # Last 3 Seasons
        last_3_seasons_points = [features.get(f'ppr_fantasy_points_s_minus_{j}', 0) for j in range(1, 4)]
        valid_points_l3 = [p for p in last_3_seasons_points if p > 0]
        if valid_points_l3:
            features['avg_ppr_last_3_seasons'] = np.mean(valid_points_l3)
            features['std_dev_ppr_last_3_seasons'] = np.std(valid_points_l3)
        else:
            features['avg_ppr_last_3_seasons'] = 0
            features['std_dev_ppr_last_3_seasons'] = 0

        # --- Year-over-Year Trends ---
        s1_points = features.get('ppr_fantasy_points_s_minus_1', 0)
        s2_points = features.get('ppr_fantasy_points_s_minus_2', 0)
        if s1_points > 0 and s2_points > 0:
            features['ppr_trend_s1_vs_s2'] = s1_points - s2_points
        else:
            features['ppr_trend_s1_vs_s2'] = 0
            
        all_features.append(features)

# --- Create Final Training DataFrame ---
if all_features:
    training_df = pd.DataFrame(all_features).fillna(0) # again, fill any last missing values with 0
    
    print("\\nSuccessfully created the training DataFrame with ENHANCED features.")
    print('\\n--- Training DataFrame Info ---')
    training_df.info()
    
    print('\\n--- Sample of Training Data Columns ---')
    print(training_df.columns.tolist())

    print('\\n--- Example: Feature set for Christian McCaffrey (a player with volatility) ---')
    # display(training_df[training_df['player_name'] == 'Christian McCaffrey'])
    print(training_df[training_df['player_name'] == 'Christian McCaffrey'])
else:
    print("\\nError: No features were generated.") 
```

Now Train the Model. GradientBoost.

```python
# This code assumes 'training_df' exists from the previous cell.
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_absolute_error, r2_score
import joblib
import os
import pandas as pd

# --- Model Preparation ---

# 1. Separate Features (X) from Target (y)
# Drop non-predictive ID columns and the target variable itself from the feature set.
FEATURES_TO_DROP = ['player_id', 'player_name', 'prediction_season', 'target_ppr_points']
X = training_df.drop(columns=FEATURES_TO_DROP)
y = training_df['target_ppr_points']

# 2. Handle Categorical Features (One-Hot Encoding)
# 'position' is a text column. Convert it into numerical format so the model can use it.
# 'pd.get_dummies' creates a new binary column for each unique position (e.g., 'position_QB', 'position_RB').
X = pd.get_dummies(X, columns=['position'], prefix='pos', drop_first=True)

# 3. Split Data into Training and Testing Sets
# Use data from seasons before 2023 to train the model, and the 2023 season data to test it.
# This chronological split mimics the real-world use case of predicting a future season.
X_train = X[training_df['prediction_season'] < 2023]
y_train = y[training_df['prediction_season'] < 2023]
X_test = X[training_df['prediction_season'] == 2023]
y_test = y[training_df['prediction_season'] == 2023]

print("Data split into training and testing sets:")
print(f"  Training samples: {len(X_train)}")
print(f"  Testing samples:  {len(X_test)}")


# --- Model Training ---
print("\nTraining the Gradient Boosting Regressor model...")

# Initialize the model with some robust starting parameters.
# n_estimators: number of boosting stages to be run.
# max_depth: maximum depth of the individual regression estimators.
# learning_rate: shrinks the contribution of each tree.
# subsample: fraction of samples to be used for fitting the individual base learners.
# random_state: for reproducibility.
gbr = GradientBoostingRegressor(
    n_estimators=500, 
    max_depth=5, 
    learning_rate=0.05, 
    subsample=0.8,
    random_state=42
)

# Train the model on the training data
gbr.fit(X_train, y_train)

print("Model training complete.")


# --- Model Evaluation ---
print("\nEvaluating model performance on the unseen 2023 test set...")

test_predictions = gbr.predict(X_test)
mae = mean_absolute_error(y_test, test_predictions)
r2 = r2_score(y_test, test_predictions)

print(f"  Mean Absolute Error (MAE): {mae:.2f}")
print(f"  R-squared (R²): {r2:.3f}")
print("\n(MAE - on average, how many fantasy points the prediction was off by)")
print("(R² - what percentage of the variance in the target variable the model explains)")


# --- Save the Trained Model ---
# save the trained model to easily load and use it later for 2024 predictions
MODEL_DIR = '../modeling/'
if not os.path.exists(MODEL_DIR):
    os.makedirs(MODEL_DIR)

model_filename = os.path.join(MODEL_DIR, 'fantasy_football_predictor.joblib')
joblib.dump(gbr, model_filename)

print(f"\nModel saved to: {os.path.abspath(model_filename)}") 
```
